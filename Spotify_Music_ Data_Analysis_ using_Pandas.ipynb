## Exploratory Data Analysis (EDA)

Exploratory Data Analysis was performed using Pandas to identify patterns, trends, and relationships within the Spotify dataset.

The analysis focused on:
- Most popular songs
- Least popular songs
- Top artists by popularity
- Genre-wise energy analysis
- Genre-wise danceability analysis
- Album popularity analysis
- Correlation between audio features

### Top 10 Most Popular Songs

```python
# Find Top 10 Most Popular Songs   
spotify[['track_name', 'artist_name', 'track_popularity']].sort_values(by='track_popularity', ascending=False).head(10)  
```

### Top 10 Least Popular Songs

```python
 # Find least 10 most popular songs 
spotify[['track_name','artist_name','track_popularity']].sort_values(by='track_popularity', ascending=False).reset_index().tail(10)        
```

## Top artists by popularity

```python
spotify.groupby('artist_name')['track_popularity'].mean().sort_values(ascending=False).head(10).reset_index()
```

This analysis identifies the artists with the highest average popularity scores in the dataset.

## Top 10 least artist

```python
# Bottom 10 artists appear in the dataset.
spotify['artist_name'].value_counts().tail(10).reset_index()       
```

This analysis identifies the artists with the least average popularity scores in the dataset.

## Top 10 Genre-wise energy analysis

```python
# Which genres are most energetic and intense.
spotify.groupby('playlist_genre')['energy'].mean().sort_values(ascending=False).reset_index().head(10)          
```

This analysis shows which music genres have the highest average energy levels.

## Top 10 Genre-wise danceability analysis


```python
# Which music genres are most danceable on average.
spotify.groupby('playlist_genre')['danceability'].mean().sort_values(ascending=False).head(10).reset_index()        
```

This analysis identifies which music genres are most danceable on average.

## Album popularity analysis


```python
spotify.groupby('track_album_name')['track_popularity'].mean().sort_values(ascending=False).head(10).reset_index()
```

This analysis highlights the albums with the highest average popularity scores.

## Top 10 Artist 

```python
# Top 10 artist appear most frequently in the dataset.
spotify['artist_name'].value_counts().head(10).reset_index()
```

These analysis highlights top 10 artist from the dataset

## Top 10 Albums by Popularity

```python
# Top 10 albums 
spotify.groupby('track_album_name')['track_popularity'].mean().sort_values(ascending=False).head(10).reset_index()  
```

This analysis identifies the top 10 albums with the highest average popularity in the dataset.

## Number of genre present in data

```python
# no. of genre present in data and which genre appeared most
spotify['playlist_genre'].value_counts().reset_index()         
```

This analysis show the number of genre present in the dataset.

## Top 10 genre by popularity

```python
# Which music genre has the highest average popularity.
spotify.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False).reset_index().head(10)         
```

This analysis shows the top 10 genre by popularity.

## Top 10 Genre for Dance

```python
# Which genres are most suitable for dancing.
spotify.groupby('playlist_genre')['danceability'].mean().sort_values(ascending=False).reset_index().head(10)         
```

This analysis shows top 10 genre for best for Dance.

## Correlation between columns

```python
spotify.corr(numeric_only=True)         # How strongly numeric columns are related to each other.
```

## Top 10 genre with the most positive / happy emotional vibe.

```python
# Top 10 genres have the most positive/happy emotional vibe.
spotify.groupby('playlist_genre')['valence'].mean().sort_values(ascending=False).reset_index().head(10)      
```

## Key Insights

- Energetic songs are generally louder.
- Danceable songs tend to have more positive emotional tones.
- Acoustic songs are usually less energetic.
- Instrumental songs often have lower loudness.
- Popularity does not strongly depend on audio features alone.
